```python
WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, "//tr[@class ='el-table__row']/td[not(contains(@class,'is-hidden'))][6]/div[not(.//button)]")))

found_time = False

page = 0

while found_time == False:
    
    try:
    
        time.sleep(1)
        
        WebDriverWait(driver, 60).until(EC.element_to_be_clickable((By.XPATH, "//td[not(contains(@class,'is-hidden'))]//button[@class='el-button el-button--default el-button--small is-round']")))

        business_name = driver.find_elements(By.XPATH, "//tr[@class ='el-table__row']/td[not(contains(@class,'is-hidden'))][6]/div[not(.//button)]")

        business_id = driver.find_elements(By.XPATH, "//tr[@class ='el-table__row']/td[not(contains(@class,'is-hidden'))][2]/div[not(.//button)]")

        update_time = driver.find_elements(By.XPATH, "//tr[@class ='el-table__row']/td[not(contains(@class,'is-hidden'))][3]/div[not(.//button)]")


        business_name_text = []

        business_id_text = []

        update_time_text = []

        for names in business_name:
                business_name_text.append(names.text)

        for ids in business_id:
                business_id_text.append(ids.text)

        for times in update_time:
                update_time_text.append(times.text)


        for index, update_time in enumerate(update_time_text):

            if update_time in time_done:
                
                found_time = True 
                
                break

            else:

                if business_name_text[index].lower().startswith("pt"):

                    temp = business_name_text[index].replace(".", " ")

                    temp_checker.append(temp)

                    print(temp)

                    business_name_list.append(business_name_text[index])

                    business_id_list.append(business_id_text[index])

                    update_time_list.append(update_time_text[index])
                    
    except:
        
        pass
                    
    try:
        
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//button[@class='btn-next']"))).click()
        
        page = page +1
        
        print('Page: '+ str(page))

        time.sleep(5)
        
    except:
        
        break
```

```python
pd.options.mode.chained_assignment = None
result = []

try:
    
    for index, all_items in checked_df['Business Name'].iteritems():

        try:

            WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, "//img[@id='logo_djahu']")))

        except:

            break

        all_items = re.sub('[^A-Za-z0-9]+', ' ', all_items)
        web_name = all_items.replace(" ", "+")
        

        change_tbk = re.compile(re.escape('tbk'), re.IGNORECASE)
        web_name = change_tbk.sub('', web_name)


        driver.get("https://ahu.go.id/pencarian/bakum/cari/tipe/perseroan?nama=" + web_name)
        

        time.sleep(1)


        full_name = driver.find_elements(By.XPATH, "//span[@class='match']/..")

        match = driver.find_elements(By.XPATH, "//span[@class='match']")

        if len(match) == 0:

            result.append('Not Match')

            checked_df['Result'][index] = 'Not Match'

            print(all_items + ' : Not Match')

            continue

        else:

            full_name = full_name[0].text.replace("PT ", "")

            if match[0].text == full_name:

                result.append('Match')

                checked_df['Result'][index] = 'Match'

                print(all_items + ' : Match')

                continue


            elif match[0].text != full_name:

                result.append('Partial Match')

                checked_df['Result'][index] = 'Partial Match'

                print(all_items + ' : Partial Match')

                continue

        time.sleep(1)
        
except:
    
    pass
```
